```
# add binary values to genres for multi label classification 
# 

drama = 'Drama'
comedy = 'Comedy'
romance = 'Romance'
action = 'Action'
crime = 'Crime'
thriller = 'Thriller'
horror = 'Horror'
adventure = 'Adventure'
documentary = 'Documentary'
mystery = 'Mystery'
family = 'Family'
fantasy = 'Fantasy'
# sci-fi = 'Sci-Fi'
biography = 'Biography'
animation = 'Animation'
history = 'History'
music = 'Music'
war = 'War'
short = 'Short'
western = 'Western'
musical = 'Musical'
sport = 'Sport'
# film-noir = 'Film-Noir'



df.insert(3, drama, 0)
df.insert(4, comedy, 0)
df.insert(5, romance, 0)
df.insert(6, action, 0)
df.insert(7, crime, 0)
df.insert(8, thriller, 0)
df.insert(9, horror, 0)
df.insert(10, adventure, 0)
df.insert(11, documentary, 0)
df.insert(12, mystery, 0)
df.insert(13, family, 0)
df.insert(14, fantasy, 0)
df.insert(15, 'Sci-Fi', 0)
df.insert(16, biography, 0)
df.insert(17, animation, 0)
df.insert(18, history, 0)
df.insert(19, music, 0)
df.insert(20, war, 0)
df.insert(21, short, 0)
df.insert(22, western, 0)
df.insert(23, musical, 0)
df.insert(24, sport, 0)
df.insert(25, 'Film-Noir', 0)
```

```
# sieve through each row and add in the appropiate binary multi-label classifications 

for index_label, row_series in df.iterrows():
  each_movie_genres = df.at[index_label, 'genre']
  array_of_genres = each_movie_genres.split('|')

  for genre in array_of_genres:

    if genre == 'Sci-Fi':
      df.at[index_label, 'Sci-Fi'] = 1
    
    if genre == 'Film-Noir':
      df.at[index_label, 'Film-Noir'] = 1

    if genre == drama:
      df.at[index_label, drama] = 1
    
    if genre == comedy:
      df.at[index_label, comedy] = 1

    if genre == romance:
      df.at[index_label, romance] = 1

    if genre == action:
      df.at[index_label, action] = 1

    if genre == crime:
      df.at[index_label, crime] = 1

    if genre == thriller:
      df.at[index_label, thriller] = 1

    if genre == horror:
      df.at[index_label, horror] = 1

    if genre == adventure:
      df.at[index_label, adventure] = 1

    if genre == documentary:
      df.at[index_label, documentary] = 1

    if genre == mystery:
      df.at[index_label, mystery] = 1

    if genre == family:
      df.at[index_label, family] = 1
    
    if genre == fantasy:
      df.at[index_label, fantasy] = 1

    if genre == biography :
      df.at[index_label, biography] = 1

    if genre == animation:
      df.at[index_label, animation] = 1

    if genre == history:
      df.at[index_label, history] = 1

    if genre == music:
      df.at[index_label, music] = 1

    if genre == war:
      df.at[index_label, war] = 1

    if genre == short:
      df.at[index_label, short] = 1

    if genre == western:
      df.at[index_label, western] = 1

    if genre == musical:
      df.at[index_label, musical] = 1
    
    if genre == sport:
      df.at[index_label, sport] = 1

df.head()
```

```
#separate dataset into training and test --- (validation?) 
import numpy as np
from sklearn.model_selection import train_test_split
#from sklearn.datasets import make_multilabel_classification

# this will generate a random multi-label dataset
# X, y = make_multilabel_classification(sparse = True, n_labels = 20,
# return_indicator = 'sparse', allow_unlabeled = False)
# print(X)
# print(y)

X_df = df.iloc[:,3:26]
y_df = df.iloc[:,26:]

# X = X_df.to_numpy()
# y = y_df.to_numpy()

X = np.asarray(df.iloc[:, 26:], dtype=np.float32)
y = np.asarray(df.iloc[:, 3:26], dtype=np.float32)
# # Todo -  repeat it for multiple times?

(X_train, X_test, Y_train, Y_test) = train_test_split(X, y, train_size=0.7, test_size=0.3)

print(X_train)
print(Y_train)

print(X_test.shape)
print(Y_test.shape)
```

### One vs Rest Classifier - SVC

```
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import hamming_loss
from sklearn.metrics import f1_score

clf = OneVsRestClassifier(SVC()).fit(X_train, Y_train)
predictions = clf.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)

print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")

```

### One vs Rest Classifier - NavieBayes 

```
from sklearn.multiclass import OneVsRestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import hamming_loss
from sklearn.metrics import f1_score

clf = OneVsRestClassifier(GaussianNB()).fit(X_train, Y_train)
predictions = clf.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)

print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")
```

### One vs Rest - Logistic Regression

### Random Forest Classifier

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import hamming_loss
from sklearn.metrics import f1_score

clf = RandomForestClassifier(max_depth=50).fit(X_train, Y_train)
predictions = clf.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)

print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")
```

### Neural Networks - MLP Classifier 

```
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import hamming_loss
from sklearn.metrics import f1_score

clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, Y_train)
predictions = clf.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)

print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")
```

### k Nearest Neighbors 

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import hamming_loss
from sklearn.metrics import f1_score

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, Y_train)
predictions = knn_clf.predict(X_test)

h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)

print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")
```

### Multi-Label kNN

```
# http://scikit.ml/api/skmultilearn.adapt.mlknn.html
from skmultilearn.adapt import MLkNN
from sklearn.metrics import accuracy_score


classifier = MLkNN(k=10)

# train
classifier.fit(X_train, Y_train)

# predict
predictions = classifier.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)
a_score = accuracy_score(Y_test,predictions)

print(f"Accuracy Score: {a_score}")
print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")

```

### Binary Relevance - Gaussian NB

```
# https://www.analyticsvidhya.com/blog/2017/08/introduction-to-multi-label-classification/
!pip install scikit-multilearn
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import hamming_loss


classifier = BinaryRelevance(GaussianNB())
classifier.fit(X_train, Y_train)

# print(Y_test)
# print(predictions)
predictions = classifier.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)
a_score = accuracy_score(Y_test,predictions)

print(f"Accuracy Score: {a_score}")
print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")

```

### Classifier Chains - Gaussian NB

```
## https://www.analyticsvidhya.com/blog/2017/08/introduction-to-multi-label-classification/
# using classifier chains
from skmultilearn.problem_transform import ClassifierChain
from sklearn.naive_bayes import GaussianNB

# initialize classifier chains multi-label classifier
# with a gaussian naive bayes base classifier
classifier = ClassifierChain(GaussianNB())

# train
classifier.fit(X_train, Y_train)

# predict
predictions = classifier.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)
a_score = accuracy_score(Y_test,predictions)

print(f"Accuracy Score: {a_score}")
print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")
```

### Label Powerset - Gaussian NB 

```
# # https://www.analyticsvidhya.com/blog/2017/08/introduction-to-multi-label-classification/

# using Label Powerset
from skmultilearn.problem_transform import LabelPowerset
from sklearn.naive_bayes import GaussianNB

# initialize Label Powerset multi-label classifier
# with a gaussian naive bayes base classifier
classifier = LabelPowerset(GaussianNB())

# train
classifier.fit(X_train, Y_train)

# predict
predictions = classifier.predict(X_test)
h_loss = hamming_loss(Y_test, predictions)
f1 = f1_score(Y_test, predictions, average=None)
a_score = accuracy_score(Y_test,predictions)

print(f"Accuracy Score: {a_score}")
print(f"Hamming loss: {h_loss}")
print(f"F1 Score: {f1}")
```
