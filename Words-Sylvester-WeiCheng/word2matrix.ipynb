```
imdbId = df6.loc[:,'imdbId']
imdbId.head()
imdbId# = imdbId.head(500)

features = df6.loc[:,'Angry':'v']
features# = features.head(500)

concat = pd.concat([imdbId, features], axis=1)

features = concat.set_index("imdbId")

print(features.shape)
print(word_matrix_words_only.shape)
# word_matrix_words_only
# features = pd.merge(features, bagofwords, on='imdbId')
features = pd.concat([features, word_matrix_words_only],join='inner',ignore_index=False, axis=1)
features = features.dropna()
features
```

```
features_id_list = features.index.values.tolist()
features_id_dict = {}
for i in features_id_list:
  features_id_dict[i] = 1

for index in list(set(outcomes.index.values.tolist())):
  if index not in features_id_dict.keys():
    outcomes = outcomes.drop([index])

outcomes
```

```
#train
from sklearn.model_selection import train_test_split
np.random.seed(782326)
x_train, x_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.3)
```

```
x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values

print(x_train.shape, y_train.shape)
```

```
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.ensemble import AdaBoostClassifier as AdaBoost
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import hamming_loss
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
import pickle
```

```
from sklearn.naive_bayes import GaussianNB
nb = OneVsRestClassifier(GaussianNB(), n_jobs=-1)
nb.fit(x_train, y_train)
pickle.dump(nb, open('nb.sav', 'wb'))
nb_pred = nb.predict(x_test)
hl = hamming_loss(y_test, nb_pred)
f1 = f1_score(y_test, nb_pred, average='micro')
print("Naive Bayes")
print("Hamming Loss of NB with words as features is " + str(hl))
print("F1 score of NB with words as features is " + str((f1)))
```
